```python
# Import Dependencies
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Import haul cycle info


haul_prfl = "Resources\MSN_lom_Profiles.xlsx"



haul_profile_df = pd.read_excel(haul_prfl)

haul_profile_df
```

```python
# Clean profile sheet Drop NUll Values

haul_profile_df.dropna(inplace=True)

haul_profile_df
```

```python
# Create ore waste leach code OWL
owl = []

for x in haul_profile_df['Dump']:
    if x == "CR1":
        y = 1
    elif x=="CR3":
        y = 1
    else:
        y= 0
    owl.append(y)

haul_profile_df["OWL"] = owl
haul_profile_df
```

```python
# Groupby PB, Bench, Dump

profiles_grouped = haul_profile_df.groupby(by=['PB','Bench',"Dump",'OWL','Dump_Z']).sum()

profiles_grouped.reset_index(inplace=True)

profiles_grouped
```

```python
# Calc grouped average lift, drop, and dist

profiles_grouped['V Lift (+)'] = profiles_grouped['lift_wt']/profiles_grouped['Tons']
profiles_grouped['V Lift (-)'] = profiles_grouped['drop_wt']/profiles_grouped['Tons']
profiles_grouped['H Dist'] = profiles_grouped['dist_w']/profiles_grouped['Tons']

profiles_grouped['int'] =  profiles_grouped['V Lift (+)'] + profiles_grouped['Bench']

profiles_grouped
```

```python
profiles_ore = profiles_grouped[profiles_grouped['OWL']==1]

profiles_waste = profiles_grouped[profiles_grouped['OWL']==0]

profiles_ore
```

```python
profiles_waste=profiles_waste.groupby(by=['PB','Bench','OWL']).sum()

profiles_waste.reset_index(inplace=True)

profiles_waste['Dump'] = "Waste"

profiles_waste
```

```python
# Calc grouped average lift, drop, and dist

profiles_waste['V Lift (+)'] = profiles_waste['lift_wt']/profiles_waste['Tons']
profiles_waste['V Lift (-)'] = profiles_waste['drop_wt']/profiles_waste['Tons']
profiles_waste['H Dist'] = profiles_waste['dist_w']/profiles_waste['Tons']

profiles_waste['int'] =  profiles_waste['V Lift (+)'] + profiles_waste['Bench']

profiles_waste['Dump_Z'] = profiles_waste['Bench'] + profiles_waste['V Lift (+)'] - profiles_waste['V Lift (-)']

profiles_waste
```

```python
dfs = [profiles_ore,profiles_waste]

profiles_cat = pd.concat(dfs)
profiles_cat
```

```python
profiles_export = profiles_cat.drop(columns=['lift_wt', 'drop_wt', 'dist_w','Tons', 'V Lift (+)', 'V Lift (-)'], axis=1)

profiles_export
```

```python
profiles_export = profiles_export.reindex(columns=['OWL','PB', 'Bench', 'Dump', 'Dump_Z','int','H Dist'])

profiles_export
```

```python
file_path = "Resources\profiles_lom_expt.csv"

profiles_export.to_csv(file_path, index=False)
```
